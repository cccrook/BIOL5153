```python
#dictionary is very similar to list except can give key instead of index

#Restrictions on dictionaries:
    #only type of data that can be used as keys are strings and numbers
    #keys must be unique (cannot have multiple of same key with different value)
    
#Most times you won't create a dictionary all in one go, you will create empty then slowly add to it

#Create empty dictionary with curly brackets all on their own
enzymes = {}

#add elements to a dictionary using square brackets on the left hand side of an assignment
enzymes['EcoRI'] = r'GAATTC'
enzymes['AvaII'] = r'GG(A|T)CC'
enzymes['BisI'] = r'GC[ATGC]GC'

#can delete from dictionary using pop method
#pop returns value and deletes key at same time

#remove EcoRI from dictionary
enzymes.pop('EcoRI')


#store the trinucleotides and their counts in a dictionary
dna = 'AATGATCGATCGTACGCTGA'
counts = {}

for base1 in ['A', 'T', 'G', 'C']:
    for base2 in ['A', 'T', 'G', 'C']:
        for base3 in ['A', 'T', 'G', 'C']:
            trinucleotide = base1 + base2 + base3
            count = dna.count(trinucleotide)
            if count > 0:
                counts[trinucleotide] = count
            print(counts)
            











```

```python
print(counts['TGA'])

#add if statement to only store counts that are greater than zero, see above

#now looking up counts is straightforward unless the count is zero, this will yield an error bc we excluded counts<1

#can avoid by checking for the existence of a key
if 'AAA' in counts:
    print(counts['AAA'])
    
#get method can be used to look up and takes 2nd argument which is value to return if key doesn't exist
print('count for TGA is ' + str(counts.get('TGA', 0)))
print('count for AAA is ' + str(counts.get('AAA', 0)))
print('count for GTA is ' + str(counts.get('GTA', 0)))
print('count for TTT is ' + str(counts.get('TTT', 0)))
```

```python
#iterating over a dictionary
#used to do something for all items in a dictionary
#works by reading list of keys in a dictionary by using keys method
#key method returns a list of all the keys in a dictionary

#print the keys
print(counts.keys())
```

```python
#print all the trinucleotides that appear twice
for trinucleotide in counts.keys():
    if counts.get(trinucleotide) == 2:
        print(trinucleotide)

#note that dictionaries are unordered-can't rely on processing items in the order you added them
#to control the order use the sorted method

#print all trinucleotides that appear twice in a sorted manner
for trinucleotide in sorted(counts.keys()):
    if counts.get(trinucleotide) == 2:
        print(trinucleotide)
```

```python
#iterating over values
#use items method
#more concise then looking up the key and then getting the value and iterating over

#less concise
#for key in my_dict.keys():
    #value = my_dict.get(key)
    #do something with key and value
    
#better way
#for key, value in my_dict.items():
    #do something with key and value

#items method returns a list of pair values as opposed to a single value or a list of values so we give two variable
#names at the start of the loop

#use item method to process dictionary of trinucleotides
for trinucleotide, count in counts.items():
    if count == 2:
        print(trinucleotide)
        







```

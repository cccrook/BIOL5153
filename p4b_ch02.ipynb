```python
#Exercise 1-Calculating AT content

#Calculate the AT content of the DNA sequence below
#ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT

dna_seq = "ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT"

#Count the occurances of 'A'
a_content = dna_seq.count('A')

#Count the occurances of 'T'
t_content = dna_seq.count('T')

#Add occurances of 'A' to occurances of 'T'
at_together = a_content + t_content

#Find the length of the DNA sequence
dna_length = len(dna_seq)

#Divide 'A + T' by length of sequence
at_content = at_together/dna_length
print(at_content)
```

```python
#Exercise 2-Complementing DNA

#Write a program that will print the complement of the DNA sequence below
#ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT

dna_seq2 = "ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT"

#Replace 'A' with 't'
new_seq = dna_seq2.replace('A', 't')

#Replace 'C' with 'G'
new_seq2 = new_seq.replace('C', 'g')

#Replace 'T' with 'A'
new_seq3 = new_seq2.replace('T', 'a')

#Replace 'G' with 'C'
fin_seq = new_seq3.replace('G', 'c')

#Print the final sequence (Note: printed 3' to 5')
print(fin_seq.upper())
```

```python
#Exercise 3-Restriction fragment lengths

#The sequence below contains a recognition site for the EcoRI restriction enzyme which cuts at the motif G*AATTC
#Write a program that will calculate the size of the two fragments it will produce from the following DNA sequence
#ACTGATCGATTACGTATAGTAGAATTCTATCATACATATATATCGATGCGTTCAT

dna_seq = "ACTGATCGATTACGTATAGTAGAATTCTATCATACATATATATCGATGCGTTCAT"

#Find the length of the sequence
length = len(dna_seq)
print('The length of the DNA sequence is:' + str(length))

#Find the recognition site
rec_seq = 'GAATTC'
rec_site = dna_seq.find(rec_seq)
print('The index of the recognition site is:' + str(rec_site))

#Split the string
frag1 = dna_seq[0:rec_site+1]
print('The sequence of the first fragementment is:' + frag1)

frag2 = dna_seq[rec_site+1:55]
print('The sequence of the second fragment is:' + frag2)

#Calculate the lengths of the two resulting strings
print('The length of the first fragement is:' + str(len(frag1)))
print('The length of the second fragment is:' + str(len(frag2)))

#Double check that fragment sizes make sense
print('The length of the two fragments should equal the length of the total sequence')
print('The length of the two fragments together is:' + str(len(frag1)+len(frag2)))

#Solution from textbook
frag1_length = dna_seq.find('GAATTC') + 1
frag2_length = len(dna_seq) - frag1_length
print('length of fragment 1 is:' + str(frag1_length))
print('length of fragment 2 is:' + str(frag2_length))
```
